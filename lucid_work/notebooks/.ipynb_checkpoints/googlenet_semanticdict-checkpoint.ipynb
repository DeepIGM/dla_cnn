```python
layer_spritemap_sizes = {
    'mixed3a' : 16,
    'mixed3b' : 21,
    'mixed4a' : 22,
    'mixed4b' : 22,
    'mixed4c' : 22,
    'mixed4d' : 22,
    'mixed4e' : 28,
    'mixed5a' : 28,
  }

def googlenet_spritemap(layer):
    assert layer in layer_spritemap_sizes
    size = layer_spritemap_sizes[layer]
    url = "https://storage.googleapis.com/lucid-static/building-blocks/googlenet_spritemaps/sprite_%s_channel_alpha.jpeg" % layer
    return size, url
```

```python
googlenet = models.InceptionV1()
googlenet.load_graphdef()


def googlenet_semantic_dict(layer, img_url):
    img = load(img_url)
    
    # Compute the activations
    with tf.Graph().as_default(), tf.Session():
        t_input = tf.placeholder(tf.float32, [224, 224, 3])
        T = render.import_model(googlenet, t_input, t_input)
        acts = T(layer).eval({t_input: img})[0]
    
    # Find the most interesting position for our initial view
    max_mag = acts.max(-1)
    max_x = np.argmax(max_mag.max(-1))
    max_y = np.argmax(max_mag[max_x])
    
    #return acts, max_mag, max_x, max_y
    
    # Find appropriate spritemap
    spritemap_n, spritemap_url = googlenet_spritemap(layer)
    
    # Actually construct the semantic dictionary interface
    # using our *custom component*
    lucid_svelte.SemanticDict({
        "spritemap_url": spritemap_url,
        "sprite_size": 110,
        "sprite_n_wrap": spritemap_n,
        "image_url": _image_url(img),
        "activations": [[[{"n": float(n), "v": float(act_vec[n])} for n in np.argsort(-act_vec)[:4]] for act_vec in act_slice] for act_slice in acts],
        "pos" : [float(max_y), float(max_x)]
    })
    
    return acts
```

```python
acts = googlenet_semantic_dict("mixed4d", "https://storage.googleapis.com/lucid-static/building-blocks/examples/dog_cat.png")
```

```python
activations = [[[{"n": float(n), "v": float(act_vec[n])} for n in np.argsort(-act_vec)[:4]] for act_vec in act_slice] for act_slice in acts]
spritemap_n, spritemap_url = googlenet_spritemap('mixed4d')
sprite_size = 110
spritemap_wrap = spritemap_n
```

```python
def get_pos_activations(pos):
    x = pos[0]
    y = pos[1]
    present_acts = activations[x][y]
    return present_acts

def get_sprite_dims(acts):
    sprites = []
    for s in acts:
        n = s['n']
        v = s['v']
        x = int(n % spritemap_wrap) * sprite_size
        y = math.floor(n/spritemap_wrap) * sprite_size
        dim = [x, y]
        val = [dim, v]
        sprites.append(val)
    return sprites
```

```python
def load_sprites(s_dims):
    sprites = []
    for i in range(4):
        value = s_dims[i][1]
        left = s_dims[i][0][0]
        upper = s_dims[i][0][1]
        right = left + sprite_size
        lower = upper + sprite_size
        crop_rectangle = (left, upper, right, lower)
        cropped_im = img.crop(crop_rectangle)
        val = cropped_im, value
        sprites.append(val)
    return sprites
        
def display_sprites(sprites):
    for i in range(len(sprites)):
        print(sprites[i][1])
        display(sprites[i][0])   
```

```python
response = requests.get(spritemap_url)
img = Image.open(BytesIO(response.content))

p_acts = get_pos_activations([0,0])
s_dims = get_sprite_dims(p_acts)
sprites = load_sprites(s_dims)
display_sprites(sprites)
```
